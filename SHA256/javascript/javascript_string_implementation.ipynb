```javascript
function bprint(input, perLine = 4) {
    let resString = "";
    let str = "";
    let len = input.length;
    let c = 1;
    while (c - 1 !== len) {
        str += input[c - 1];
        if (c && c%4 === 0) {
            resString += str + " ";
            if (resString.split(' ').length - 1 === perLine) {
                console.log(resString);
                resString = "";
            }
            str = "";
        }
        c += 1;
    }
}
```

```javascript
function pad(binaryArray) {
    let output = binaryArray;
    let rightl = binaryArray.length.toString(2);
    let flen = 512 - rightl.length;
    output += "10";
    while (output.length%flen !== 0) {
        output += "0";
    }
    return output + binaryArray.length.toString(2);
}
```

```javascript
function convToWordArray(input) {
    let res = [];
    for (let i in input) {
        i = parseInt(i)
        if ((i)%32 === 0) {
            res.push(input.slice(i, i + 32));
        }
    }
    return res;
}
```

```javascript
shiftByAmount = (str, leftShifts) => {
    // helper function
    // negative amount shifts to right
    // positive amount shifts to left
    leftShifts = leftShifts % str.length;
    return str.slice(leftShifts) + str.slice(0, leftShifts);
};

shiftRightNotCircular = (str, n) => {
    const circus = Array(n).fill(0).join("");
    return circus + str.slice(0, str.length - n);
}

logic = {
    RotR: (A, n) => shiftByAmount(A.toString(), -n),
    ShR: (A, n) => shiftRightNotCircular(A.toString(), n),
    concat: (A, B) => `${A}${B}`,
}
package = {
    Ch: (x, y, z) => (x&y)^((!x)&z),
    Maj: (x, y, z) => (x&y)^(x&z)^(y&z),
    E0: (x) => logic.RotR(x,2)^logic.RotR(x,13)^logic.RotR(x,22),
    E1: (x) => logic.RotR(x,6)^logic.RotR(x,11)^logic.RotR(x,25),
    e0: (x) => logic.RotR(x,7)^logic.RotR(x,18)^logic.ShR(x,3),
    e1: (x) => logic.RotR(x,17)^logic.RotR(x,19)^logic.ShR(x,10)
}
```

```javascript
function convBinToInts(arr) {
    return arr.map((i) => parseInt(i, 2));
}

function schedule(wow) {
    const w = convBinToInts(wow);
    for (let i = 16; i <= 63; i++) {
        /* 
            s0 = (w[i-15] rightrotate 7) xor (w[i-15] rightrotate 18) xor (w[i-15] rightshift 3)
            s1 = (w[i- 2] rightrotate 17) xor (w[i- 2] rightrotate 19) xor (w[i- 2] rightshift 10)
            w[i] = w[i-16] + s0 + w[i-7] + s1
        */
        s0 = (logic.RotR(w[i-15], 7))^(logic.RotR(w[i-15], 18))^(logic.ShR(w[i-15], 3));
        s1 = (logic.RotR(w[i-2], 17))^(logic.RotR(w[i-2], 19))^(logic.ShR(w[i-15], 10));
        w[i] = w[i-16] + s0 + w[i-7] + s1;
    }
    return w;
}

sch = schedule(wordArr).map(i => i.toString(2));
```
